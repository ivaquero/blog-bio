### 4.1. Cohen's _d_

Jacob Cohenï¼ˆ1932âˆ¼1998ï¼‰å®šä¹‰äº† _d_ å€¼å’Œä¸¤ä¸ªæ€»ä½“çš„å¹¶åˆæ ‡å‡†åå·®ï¼ˆpooled standard deviationï¼Œ$S_p^2$ï¼‰

$$
d = \frac{XÌ„_1 - XÌ„_2}{S_p}
$$

ç”± _t_ å€¼å®šä¹‰ï¼Œ$d = t/\sqrt n$ï¼Œå¯¹$k$ä¸ªä¸åŒæ€»ä½“çš„æ ·æœ¬

$$
S_p^2 = \frac{âˆ‘_{i=1}^{k} (n_{i} -1)S_{i}^2}{âˆ‘_{n - 1}^{k} (n_{i} -1)}
$$

å…¶ä¸­ï¼Œ$S_{i}^2$ä¸ºæ¯ä¸ªæ ·æœ¬çš„æ–¹å·®ã€‚

å½“$k = 2$æ—¶ï¼Œ$S_p^2$å³ä¸ºä¸åŒæ€»ä½“ _t_ æ£€éªŒä¸­çš„$S_w^2$ï¼Œå³æœ‰

$$
d = \frac{D}{Ïƒ}
$$

2009 å¹´ï¼ŒShlomo S. Sawilowskyï¼ˆ1954âˆ¼ï¼‰åœ¨ Cohen çš„åŸºç¡€ä¸Šï¼Œå®Œå–„äº†å¯¹ _d_ å€¼æ•ˆåº”çš„æè¿°ã€‚

### 4.3. æ ·æœ¬é‡

è®¾ä»Žä¸­æŠ½å–æ ·æœ¬çš„æ€»ä½“çš„å‡å€¼ä¸º$XÌ‚$ä¸”æ ‡å‡†å·®ä¸º Ïƒï¼Œè€Œå®žé™…æ€»ä½“çš„å‡å€¼ä¸º$Î¼ = XÌ‚ + D$ä¸”æ ‡å‡†å·®ç›¸åŒï¼Œåˆ™æœ€å°æ ·æœ¬æ•°é‡

$$
n = \frac{(z_{1 - Î±/2} + z_{1 - Î²})^2}{d^2}
$$

å…¶ä¸­ï¼Œ$z = \dfrac{x - Î¼}{Ïƒ}$ä¸º _Z_ ç»Ÿè®¡é‡ï¼Œ$d$ä¸ºæ•ˆåº”é‡ã€‚

æ¢å¥è¯è¯´ï¼Œè‹¥çœŸå®žå‡å€¼çš„å€¼ä¸º$XÌ‚$ï¼Œæ‰€æœ‰æ£€éªŒä¸­çš„è‡³å°‘$1 - Î±ï¼…$å¯æ­£ç¡®æ£€æµ‹åˆ°ï¼›è‹¥çœŸå®žå‡å€¼åç§»äº†$D$æˆ–æ›´å¤šï¼Œåˆ™åº”æ£€æµ‹åˆ°è¿™ä¸ªå¯èƒ½æ€§è‡³å°‘ä¸º$1 - Î²ï¼…$ã€‚

å¯¹ä¸åŒæ€»ä½“

$$
n_1 = n_2 = \frac{(z_{1 - Î±/2} +z_{1 - Î²})^2(Ïƒ^2_1 + Ïƒ^2_2)}{D^2}
$$


```python
def sampleSize_oneGroup(d, Î±, Î², Ïƒ=1) -> np.float16:
    n = np.round((stats.norm.ppf(1 - Î± / 2) + stats.norm.ppf(1 - Î²)) ** 2 * Ïƒ**2 / d**2)
    print(f"You need at least {int(n):d} subjects.")
    return n


def sampleSize_twoGroups(d, Î±, Î², Ïƒ1=1, Ïƒ2=1) -> np.float16:
    n = np.round(
        (stats.norm.ppf(1 - Î± / 2) + stats.norm.ppf(1 - Î²)) ** 2
        * (Ïƒ1**2 + Ïƒ2**2)
        / d**2
    )
    print(f"You need at least {int(n):d} subjects.")
    return n


Î±, Î² = 0.05, 0.2
sampleSize_oneGroup(0.5, Î±, Î²)
print("-" * 30)
sampleSize_twoGroups(0.4, Î±, Î², Ïƒ1=0.6, Ïƒ2=0.6)
```
